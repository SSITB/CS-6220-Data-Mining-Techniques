```python
def same_rating(uid, mid):
    user_ids = []
    select_row = genres_unique[(genres_unique['user_id']==uid) & (genres_unique['movie_id']==mid)]
    rting = select_row.rating[0]
    select_users = genres_unique[(genres_unique['rating']==rting) & (genres_unique['movie_id']==mid)]
    return select_users.user_id
```

```python
def sim_distance(prefs,person1,person2):
    
    # Get the list of shared_items
    si={}
    for item in prefs[person1]:
        if item in prefs[person2]:
            si[item]=1
    
    # if they have no ratings in common, return 0
    if len(si)==0: return 0
    
    # Add up the squares of all the differences
    sum_of_squares=sum([pow(prefs[person1][item]-prefs[person2][item],2)
                        for item in prefs[person1] if item in prefs[person2]])

    return 1/(1+sum_of_squares)
```

```python
sim_distance(critics,'Lisa Rose','Toby')
```
